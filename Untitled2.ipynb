```python
import requests

url = 'http://localhost:8000/chat'  # URL de la ruta del chat en tu aplicación Flask

# Mensaje de entrada para el chat
message = {
    'message': 'Como siempre, utilizaremos las librerías pulp y pandas. Adicionalmente, usaremos algunas funciones de la librería math para calcular distancias geográficas de forma rápida y una función de la librería datetime para calcular el tiempo de ejecución del optmizador.'
}

# Realizar la solicitud POST
response = requests.post(url, json=message)
data = response.json()

# Obtener la respuesta del chat
answer = data['answer']
print(answer)
```
